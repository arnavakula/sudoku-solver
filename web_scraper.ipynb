```python
from bs4 import BeautifulSoup
import requests
import re
import json
```

```python
url = 'https://www.nytimes.com/puzzles/sudoku/easy'
```

```python
soup = BeautifulSoup(requests.get(url).content,"html.parser")
```

```python
soup.find('div', {'class':'pz-game-screen'})
```

```python
str = soup.find('script').contents[0]
```

```python
str
```

```python
puzzle_data = json.loads(re.sub('window.gameData = ', '', str))
```

```python
puzzle_data
```

```python
pz = puzzle_data['easy']['puzzle_data']['puzzle']
len(pz)
```

```python
board = []
for i in range(9):
    board.append(pz[9 * i:9 * i + 9])

board
```
